```python
import numpy as np
import matplotlib.pyplot as plt

from hypercube import Hypercube

plt.rcParams['figure.figsize'] = (5, 5)

N = 6
```

```python
points = np.random.normal(size=(N, 2))
plt.scatter(points[:, 0], points[:, 1])
plt.show()
```

```python
X = points - np.min(points)
plt.scatter(X[:, 0], X[:, 1])
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.show()
```

```python
bounds = np.zeros((2, 2))
bounds[:, 1] = np.max(X)
H_0 = Hypercube(bounds, X, 0)

H_0.divide()
```
